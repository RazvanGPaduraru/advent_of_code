```python
from collections import defaultdict, deque


directions = [[-1, 0], [0, 1], [1, 0], [0, -1]]

visited_dict = dict()
directed_fence = dict()


garden = []

def is_position_inside(i, j):
    return 0 <= i < rows and 0 <= j < columns

with open('data_day12.txt', 'r') as file:
     garden = [list(line.strip()) for line in file if line.strip()]


row_length = len(garden[0])  
border_row = ['*'] * (row_length + 2)  

surrounded_garden = [border_row]  
for row in garden:
    surrounded_garden.append(['*'] + row + ['*'])  
surrounded_garden.append(border_row)  


rows = len(garden)
columns = len(garden[0])

def count_fences(fence_map):
    fences_by_direction = defaultdict(list)
    for key, value in fence_map.items():
        if value == 1:  
            position, direction = key
            fences_by_direction[direction].append(position)
    
    def find_connected_components(blocks):
        visited = set()
        components = 0
        
        def bfs(start):
            queue = deque([start])
            visited.add(start)
            while queue:
                x, y = queue.popleft()
                for dx, dy in directions:
                    neighbor = (x + dx, y + dy)
                    if neighbor in blocks and neighbor not in visited:
                        visited.add(neighbor)
                        queue.append(neighbor)
        for block in blocks:
            if block not in visited:
                components += 1
                bfs(block)
        
        return components
    total_fences = 0
    for direction, blocks in fences_by_direction.items():
        total_fences += find_connected_components(set(blocks))
    
    return total_fences



def compute_fence(i, j, plant):
    global visited_dict

    visited_dict[(i, j)] = 1
    number_of_fences = 0
    number_of_plants = 0


    for index, path in enumerate(directions):
        new_i = i + path[0]
        new_j = j + path[1]

        if not is_position_inside(new_i, new_j) or garden[new_i][new_j] != plant:
            directed_fence[((new_i, new_j), index)] = 1
            number_of_fences += 1
        elif is_position_inside(new_i, new_j) and (new_i, new_j) not in visited_dict and garden[new_i][new_j] == plant:
            number_of_plants+=1
            
            new_fences, new_plants = compute_fence(new_i, new_j, plant)

            number_of_fences += new_fences
            number_of_plants += new_plants

    return (number_of_fences, number_of_plants)


cost_of_fences = 0

for i in range(rows):
    for j in range(columns):
        if (i, j) not in visited_dict:
            directed_fence = {}
            s = compute_fence(i, j, garden[i][j])

            cost_of_fences += ((s[1] + 1) * count_fences(directed_fence))

print(cost_of_fences)   
            



```
