```python
with open('data_day9.txt', 'r') as file:
    data = file.read().replace('\n', '')
    disk_map = list(data)
    disk_files = []
    free_spaces = []
    poz_id = {}
    for i in range (len(disk_map)):
        poz_id[i] = []
        if i % 2 == 0:
            disk_files.append(int(disk_map[i]))
        else:
            free_spaces.append(int(disk_map[i]))


array = []

current_id = 0
for i in range(len(disk_files)):
    for j in range(disk_files[i]):
        array.append(current_id)
    
    if i < len(free_spaces):
        for k in range(free_spaces[i]):
            array.append('.')

    current_id +=1

current_space_index = 0 
already_moved = {}


def find_available_block(start_idx, free_space):
    for i in range(len(disk_files) - 1, start_idx, -1):
        if disk_files[i] <= free_space and i not in already_moved:
            return (i, disk_files[i])
    return (-1, -1)
        


new_array = []



for i in range(len(array)):
    new_array.append('.')

for i in range(0, len(array)):

    if(i > 0 and new_array[i] != '.' and new_array[i-1] == '.'):
        current_space_index += 1

    
    if array[i] != '.' and array[i] not in already_moved:
        new_array[i] = array[i]
    
    if new_array[i] != new_array[i-1] and new_array[i-1] != '.':
        already_moved[new_array[i-1]] = 1
    
    if new_array[i] == '.' and array[i] == '.' and current_space_index < len(free_spaces):

        j = i
        count = 0
        while (array[j] == '.'):
            count += 1
            j += 1

        available_block = find_available_block(0, count)

        if available_block != (-1, -1):

            for j in range(i, i + available_block[1]):
                new_array[j] = available_block[0]
            i = i + available_block[1] - 1
            already_moved[available_block[0]] = 1
            free_spaces[current_space_index] -= available_block[1]

            if free_space == available_block[1]:
                current_space_index += 1



            
        

s = 0
for i in range(len(new_array)):
    if new_array[i] != '.':
        s += (i * int(new_array[i]))

print(s)


```
